```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Load DataSet

```python
train = pd.read_csv("./titanic_train.csv", index_col="PassengerId")
print(train.shape)
train.head()
```

```python
test = pd.read_csv("./titanic_test.csv", index_col="PassengerId")
print(test.shape)
test.head()
```

## Explore

```python
# matplotlib에 화면에 바로 띄우기. seaborn은 matplotlib 기반
%matplotlib inline
import seaborn as sns
```

### CountPlot

```python
sns.countplot(data = train, x="Embarked", hue="Survived")
```

```python
sns.countplot(data = train, x="Pclass", hue="Survived")
```

```python
# 검은색 bar는 표준편차
sns.barplot(data=train, x="Pclass", y="Fare", hue="Survived")
```

### PointPlot

```python
# x축 data가 연관성 있을 경우 pointPlot
sns.pointplot(data=train, x="Pclass", y="Fare", hue="Survived")
```

### DistPlot

```python
# distribution
sns.distplot(train["Fare"], hist=False)
```

```python
# outlier 없애기
low_fare = train[train["Fare"] < 100]
print(train.shape, low_fare.shape)
sns.distplot(low_fare["Fare"], hist=False)
```

### LmPlot

```python
low_fare1 = train[train["Fare"] < 500]
sns.lmplot(data=low_fare1, x="Age", y="Fare", hue="Survived", fit_reg=False)
```

### 여러개

```python
train2 = pd.read_csv("./train.csv", parse_dates=["datetime"])
```

```python
train2["datetime_year"] = train2["datetime"].dt.year
train2["datetime_month"] = train2["datetime"].dt.month
train2["datetime_day"] = train2["datetime"].dt.day
train2["datetime_hour"] = train2["datetime"].dt.hour
train2["datetime_min"] = train2["datetime"].dt.minute
train2["datetime_sec"] = train2["datetime"].dt.second
train2["datetime_dayofweek"] = train2["datetime"].dt.dayofweek
```

```python
# matplotlib의 subplots를 사용합니다. 이 함수는 여러 개의 시각화를 한 화면에 띄울 수 있도록 합니다.
# 이번에는 2x3으로 총 6개의 시각화를 한 화면에 띄웁니다.
figure, ((ax1, ax2, ax3), (ax4, ax5, ax6)) = plt.subplots(nrows=2, ncols=3)

# 시각화의 전체 사이즈는 18x8로 설정합니다.
figure.set_size_inches(18, 8)

# seaborn의 barplot으로 subplots의 각 구역에
# 연, 월, 일, 시, 분, 초 별 자전거 대여량을 출력합니다.
sns.barplot(data=train2, x="datetime_year", y="count", ax=ax1)
sns.barplot(data=train2, x="datetime_month", y="count", ax=ax2)
sns.barplot(data=train2, x="datetime_day", y="count", ax=ax3)
sns.barplot(data=train2, x="datetime_hour", y="count", ax=ax4)
sns.barplot(data=train2, x="datetime_min", y="count", ax=ax5)
sns.barplot(data=train2, x="datetime_sec", y="count", ax=ax6)
```
