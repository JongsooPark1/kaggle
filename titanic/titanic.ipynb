```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
```

## load dataSet

```python
train = pd.read_csv("../day1/dataSet/train.csv", index_col="PassengerId")
print(train.shape)
train.head()
```

```python
test = pd.read_csv("../day1/dataSet/test.csv", index_col="PassengerId")
print(test.shape)
test.head()
```

## preprocessing

### Encode Sex

```python
train.loc[train["Sex"] == "male", "Sex_encode"] = 0
train.loc[train["Sex"] == "female", "Sex_encode"] = 1

print(train.shape)
train[["Sex", "Sex_encode"]].head()
```

```python
test.loc[test["Sex"] == "male", "Sex_encode"] = 0
test.loc[test["Sex"] == "female", "Sex_encode"] = 1

print(test.shape)
test[["Sex", "Sex_encode"]].head()
```

## Fill in missing fare

```python
train[train["Fare"].isnull()]
```

```python
test[test["Fare"].isnull()]
```

```python
train["Fare_fillin"] = train["Fare"]
test["Fare_fillin"] = test["Fare"]

print(train.shape)
print(test.shape)

test[["Fare", "Fare_fillin"]].head()
```

```python
test.loc[test["Fare_fillin"].isnull(), "Fare_fillin"] = 0
```

```python
columns = ["Fare", "Fare_fillin"]
test.loc[test["Fare"].isnull(), columns]
```

### Encode Embarked

```python
# One Hot Encoding 쓰는 이유
# C == 0, S == 1, Q == 2 ? 이렇게 하면 안된다. 두 가지면 그냥 0, 1로
# 2 * S = Q? -> No
# S + Q = Q? -> No...이기 떄문

# One Hot Encoding
# C == [1, 0, 0]
# S == [0, 1, 0]
# Q == [0, 0, 1]

# True == 1, False == 0
# C == [True, False, False]
# S == [False, True, False]
# Q == [False, False, True]

# Pclass도 대상이긴 한데, 적용하면 더 accuracy 떨어진다...why?

train["Embarked_C"] = train["Embarked"] == "C"
train["Embarked_S"] = train["Embarked"] == "S"
train["Embarked_Q"] = train["Embarked"] == "Q"

print(train.shape)
columns2 = ["Embarked", "Embarked_C", "Embarked_S", "Embarked_Q"]
train[columns2]
```

```python
test["Embarked_C"] = test["Embarked"] == "C"
test["Embarked_S"] = test["Embarked"] == "S"
test["Embarked_Q"] = test["Embarked"] == "Q"

print(test.shape)
test[columns2]
```

### Train

```python
feature_names = ["Pclass", "Sex_encode", "Fare_fillin", "Embarked_C", "Embarked_S", "Embarked_Q"]
feature_names
```

```python
X_train = train[feature_names]

print(X_train.shape)
X_train.head()
```

```python
X_test = test[feature_names]

print(X_test.shape)
X_test.head()
```

```python
label_name = "Survived"
y_train = train[label_name]
print(y_train.shape)
y_train.head()
```

## Use Decision Tree

```python
model = DecisionTreeClassifier(max_depth=5)
model
```

```python
# 1) fit(train, 학습) -> 2) predict(test)
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)

print(predictions.shape)
predictions[0:5]
```
